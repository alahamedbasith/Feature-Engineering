### ORDINAL NUMBER ENCODING

```python
import datetime
```

```python
# To fetch today date
today_date = datetime.datetime.today()
```

```python
today_date
```

```python
# I want to fetch the last 10 date 
today_date - datetime.timedelta(2)
```

```python
### List Comprehension
[today_date - datetime.timedelta(x) for x in range(0,15)]
```

```python
days = [today_date - datetime.timedelta(x) for x in range(0,15)]

# Convert into dataframe
import pandas as pd
data = pd.DataFrame(days)

data.head()
```

```python
data.columns = ['Day']
```

```python
data.head()
```

```python
# Extract the weekend days
data['Weekday'] = data['Day'].dt.day_name()
data.head()
```

```python
# Then create the dictionary

keys = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
values = [1,2,3,4,5,6,7]

dictionary = dict(zip(keys,values))
```

```python
dictionary
```

```python
data['Weekday'].map(dictionary)
```

```python
data['Weekday_Ordinal'] = data['Weekday'].map(dictionary)
```

```python
data.head()
```

## Count or Frequency Encoding

```python
import pandas as pd
train_data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)
```

```python
train_data.head(10)
```

```python
columns = [1,3,5,6,7,8,9,13]
# Selected columns only
train_data = train_data[columns]
train_data.columns = ['Employment','Degree','Status','Designation','Family_Job','Race','Sex','Country']
train_data.head()
```

```python
for feature in train_data.columns:
    print(feature,":",len(train_data[feature].unique()),'labels')
```

```python
# There are some labels are available we encode anything now i encoded country because it has more number of labels

train_data['Country'].value_counts()
```

```python
# to covert dictionary

dictionary_map = train_data['Country'].value_counts().to_dict()
dictionary_map
```

```python
train_data['Country'] = train_data['Country'].map(dictionary_map)
```

```python
train_data.head()
```

The values are encoded successfully

#### Advantages
1.Easy to Implement <br>
2.Not Increasing the feature space

#### Disadvantages
1.It will provides same weights if the frequencies are same
