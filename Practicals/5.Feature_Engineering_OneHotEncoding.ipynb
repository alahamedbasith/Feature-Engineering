### There is a problem of taking the all features to perform one hot encoding.It leads to cause curse of dimensionality
In order to avoid that,In kaggle KDD Orange competition authors says you need to take top 10 feature categories among the all.It also prove and Success for many use cases

```python
df['X0'].value_counts(ascending=False).head(10)
```

```python
top_10 = df['X0'].value_counts(ascending=False).head(10).index
```

```python
top_10_list = list(top_10)
```

```python
top_10_list
```

```python
import numpy as np
for categories in top_10_list:
    df[categories] = np.where(df['X0']==categories,1,0)
```

```python
df.head(10)
```

```python
df[top_10_list].head()
```

```python
top_10_list.append('X0')
```

```python
df[top_10_list].head(10)
```

## Perform one hot encoding for all of the features

```python
df = pd.read_csv('D:\\Feature Engineering\\Datasets\\Mercedes Datasets\\mercedes.csv',usecols = ['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
df.columns
```

```python
top_10 = []  # Initialize an empty list to store the top 10 values for each feature

for feature in df.columns:
    top_10_for = df[feature].value_counts().sort_values(ascending=False).head(10)
    top_10.append(top_10_for)

# Now, top_10 contains the top 10 values for each feature
```

```python
top_10
```

```python
top_10_list = list(top_10)
top_10_list
```
